```python
import keras
from keras.optimizers import Adam
import tensorflow as tf
from keras import layers
from keras.models import Sequential
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import Sequential
from tensorflow.keras.layers import (
    AveragePooling2D,
    BatchNormalization,
    Conv2D,
    Dense,
    Flatten,
    MaxPool2D,
    PReLU,
    ReLU,
    MaxPooling2D,
    Add
)

import pandas as pd
import numpy as np

tf.random.set_seed(123)
np.random.seed(123)
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(
    "/dataset/train",
    label_mode="binary",
    color_mode="rgb",
    batch_size=32,
    shuffle=True,
    seed=123,
    image_size=(224, 224))


test_ds = tf.keras.utils.image_dataset_from_directory(
    "/dataset/test",
    label_mode="binary",
    color_mode="rgb",
    batch_size=32,
    shuffle=True,
    seed=123,
    image_size=(224, 224))
```

```python
# import the model
seq = tf.keras.models.load_model("/models/cnn_01/cnn_01.keras", safe_mode = False)
```

```python
seq.compile(loss="binary_crossentropy",
              optimizer= Adam(learning_rate= 0.001),
              metrics=["accuracy"])
```

```python
history = seq.fit(train_ds, validation_data=test_ds, epochs=300)
```

```python
history_df = pd.DataFrame(history.history)
path = 'models/cnn_01/cnn_01_hist.csv'
with open(path, mode='w') as f:
    history_df.to_csv(f)
```
